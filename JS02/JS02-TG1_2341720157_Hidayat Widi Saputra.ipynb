# 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df.isnull().sum()
```

Memisahkan `id` dan `Unnamed: 32`, Karena data tidak berkaitan

```python
df_before = df
df = df.drop(['id', 'Unnamed: 32'], axis=1)
df.isnull().sum()
```

# 2. Lakukan proses encoding pada kolom "diagnosis".

```python
df.head(100)
```

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
df.head(100)
```

# 3. Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
numeric_cols.remove('diagnosis')
df[numeric_cols] = std.fit_transform(df[numeric_cols])

df.head()
```
