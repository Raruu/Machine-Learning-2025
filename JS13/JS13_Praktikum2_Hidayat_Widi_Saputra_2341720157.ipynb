Klasifikasi Data Iris


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)

# One-hot encoding
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Bangun model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

# Kompilasi
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Latih model
model.fit(X_train, y_train, epochs=50, batch_size=8)

# Evaluasi
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi: {acc}")
```

# Tugas 2

## Ubah jumlah neuron hidden layer.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)

# One-hot encoding
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Bangun model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(12, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

# Kompilasi
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Latih model
model.fit(X_train, y_train, epochs=50, batch_size=8)

# Evaluasi
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi: {acc}")
```

## Bandingkan Sigmoid vs ReLU pada dataset Iris.

## Catat perbedaan loss dan akurasi.


```python
params = [
    {
        "name": "ReLU - Konfigurasi [10, 8]",
        "layers": [10, 8],
        "activation": "relu",
    },
    {
        "name": "Sigmoid - Konfigurasi [10, 8]",
        "layers": [10, 8],
        "activation": "sigmoid",
    },
    {
        "name": "ReLU - Hidden Layer [32, 16]",
        "layers": [32, 16],
        "activation": "relu",
    },
    {
        "name": "Sigmoid - Hidden Layer [32, 16]",
        "layers": [32, 16],
        "activation": "sigmoid",
    },
]

results = []

for config in params:
    print(f"\n{config['name']}")

    model = tf.keras.Sequential()
    model.add(
        tf.keras.layers.Dense(
            config["layers"][0], activation=config["activation"], input_shape=(4,)
        )
    )

    for neurons in config["layers"][1:]:
        model.add(tf.keras.layers.Dense(neurons, activation=config["activation"]))

    model.add(tf.keras.layers.Dense(3, activation="softmax"))

    # Kompilasi
    model.compile(
        optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"]
    )

    # Latih model
    history = model.fit(
        X_train, y_train, epochs=50, batch_size=8, verbose=0, validation_split=0.2
    )

    # Evaluasi
    loss, acc = model.evaluate(X_test, y_test, verbose=0)

    # Ambil loss terakhir dari training
    final_train_loss = history.history["loss"][-1]
    final_train_acc = history.history["accuracy"][-1]

    results.append(
        {
            "Konfigurasi": config["name"],
            "Fungsi Aktivasi": config["activation"].upper(),
            "Arsitektur": str(config["layers"]),
            "Train Loss": round(final_train_loss, 4),
            "Test Loss": round(loss, 4),
            "Train Acc": round(final_train_acc, 4),
            "Test Acc": round(acc, 4),
            "Test Acc (%)": round(acc * 100, 2),
        }
    )

    print(f"Train Loss: {final_train_loss:.4f}, Test Loss: {loss:.4f}")
    print(f"Train Acc: {final_train_acc:.4f}, Test Acc: {acc:.4f} ({acc*100:.2f}%)")
```
